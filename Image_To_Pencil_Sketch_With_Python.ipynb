# Import Libraries

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

# Load the Image

```python
#imread()- This function will load the image  
img = cv2.imread("Tiger.jpg")
```

```python
# An image is just a two-dimensional array of integers. 
# So we can perform matrix manipulations using various python modules in order to get some very interesting effects
img
```

# Read the Image in RGB Format

```python
# OpenCV interprets images in BGR format
# matplotlib interprets images in RGB format
# so to convert the image so that it's properly loaded, convert it before loading

# cvtColor()- This function takes color as an argument and then changes the source image color into that color.

rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(rgb_img)
plt.show()
```

# Convert Image to Grayscale Image

```python
gray_scale_img = cv2.cvtColor(rgb_img, cv2.COLOR_RGB2GRAY)

plt.imshow(gray_scale_img)
plt.show()
```

# Inverting the Gray Scale Image(This will make it a Negative Image)

```python
# bitwise_not()- This function will help the image to keep the properties as same by providing the masking to it.
invert_img = cv2.bitwise_not(gray_scale_img)

plt.imshow(invert_img)
plt.show()
```

# Convert to Blur Image

```python
# GaussianBlur()- This function is used to modify the image by sharpening the edges of the image, 
# smoothen the image, and will minimize the blurring property.

blur_img = cv2.GaussianBlur(invert_img, (25,25),0)

plt.imshow(blur_img)
plt.show()
```

# Invert the Blur Image 

```python
invert_blur_img = cv2.bitwise_not(blur_img)

plt.imshow(invert_blur_img)
plt.show()
```

# Pencil Sketch by Mixing Grayscale image and Inverted blurry Image

```python
# divide()- This function is used for the normalization of the image as it doesn’t lose its previous properties.
sketch_img = cv2.divide(gray_scale_img,invert_blur_img,scale=256)

rgb_sketch = cv2.cvtColor(sketch_img,cv2.COLOR_BGR2RGB)

plt.imshow(rgb_sketch)
plt.show()
```

# Saving the Sketch 

```python
cv2.imwrite("sketch.png", rgb_sketch)
```

# Original Image Converted to Sketch

```python
plt.figure(figsize=(25,15))

plt.subplot(1,2,1)
plt.title('Original Image',size=30)
plt.imshow(rgb_img)
plt.axis('off')

plt.subplot(1,2,2)
plt.title('Sketch',size=30)
rgb_sketch = cv2.cvtColor(sketch_img,cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')

plt.show()
```
