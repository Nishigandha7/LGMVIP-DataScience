```python
df = df[['iyear','imonth','iday','country_txt','region_txt','provstate','city','latitude','longitude','location','summary',
         'attacktype1_txt','targtype1_txt','gname','motive','weaptype1_txt','nkill','nwound','addnotes']]
```

```python
df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','region_txt':'Region',
                   'provstate':'State','city':'City','latitude':'Latitude','longitude':'Longitude','location':'Location',
                   'summary':'Summary','attacktype1_txt':'Attack Type','targtype1_txt':'Target Type','gname':'Group Name',
                   'motive':'Motive','weaptype1_txt':'Weapon Type','nkill':'Killed','nwound':'Wounded','addnotes':'Add Notes'},
         inplace=True)
```

```python
df1 = df[['Year','Month','Day','Country','Region','State','City','Latitude','Longitude','Location','Summary',
         'Attack Type','Target Type','Group Name','Motive','Weapon Type','Killed','Wounded','Add Notes']]
```

# Finding the total number of attacks in each year (1970 - 2017)

```python
attack = df['Year'].value_counts().sort_index()
attack = attack.to_frame()
attack = attack.reset_index()
attack = attack.rename(columns={'index':'Year','Year':'Attacks'})
attack = attack.set_index('Year')
attack
```

```python
attack.plot(kind='bar',figsize = (17,10))
plt.ylabel('Number of Attacks')
plt.show()
```

# Finding the total number of killed people in attacks each year(1970-2017)

```python
killed = df[['Year','Killed']].groupby('Year').sum()
killed
```

```python
killed.plot(kind='bar',color='red',figsize = (17,10))
plt.ylabel('Number of People Killed')
plt.show()
```

# Finding the total number of wounded people in attacks each year(1970-2017)

```python
wounded = df[['Year','Wounded']].groupby('Year').sum()
wounded
```

```python
wounded.plot(kind='bar',color='lime',figsize = (17,10))
plt.ylabel('Number of Wounded People')
plt.show()
```

# Finding Region wise attacks 

```python
region = pd.crosstab(df1.Year,df1.Region)
region
```

```python
region.plot(kind='area',figsize=(17,10))
plt.ylabel('Number of Attacks')
plt.show()
```

# Finding Country wise attacks

```python
country_wise_attacks = df1['Country'].value_counts()
country_wise_attacks
```

# Finding the country with most number of attacks

```python
country_wise_attacks = country_wise_attacks.head(10)
country_wise_attacks.plot(kind='bar',color='cyan',figsize = (17,10))
plt.ylabel('Number of Attacks')
plt.show()

print("Country with the most attacks:",df1['Country'].value_counts().idxmax())
```

# Finding City wise attacks

```python
city_wise_attacks = df1['City'].value_counts()
city_wise_attacks
```

# Finding the city with most number of attacks

```python
city_wise_attacks = city_wise_attacks[1:11]
city_wise_attacks.plot(kind='bar',color='yellow',figsize = (17,10))
plt.ylabel('Number of Attacks')
plt.show()

print("City with the most attacks:",df1['City'].value_counts().index[1])
```

# Finding region wise attacks

```python
region_wise_attacks = df1['Region'].value_counts()
region_wise_attacks
```

# Finding the region with most number of attacks 

```python
region_wise_attacks.plot(kind='bar',color='pink',figsize = (17,10))
plt.ylabel('Number of Attacks')
plt.show()

print("Region with the most attacks:",df1['Region'].value_counts().idxmax())
```

# Finding month wise attacks 

```python
month_wise_attack = df1['Month'].value_counts()
month_wise_attack
```

# Finding the month with most number of attacks

```python
month_wise_attack.plot(kind='bar',color='green',figsize = (17,10))
plt.ylabel('Number of Attacks')
plt.show()

print("Month with the most attacks:",df1['Month'].value_counts().idxmax())
```

# Finding the total number of attacks by different terrorist groups

```python
terrorist_group = df1['Group Name'].value_counts()
terrorist_group
```

# Finding the terrorist group with most number of attacks

```python
terrorist_group = terrorist_group[1:11]
terrorist_group.plot(kind='bar',color='red',figsize = (17,10))
plt.ylabel('Number of Attacks')
plt.show()

print("Group with the most attacks:",df1['Group Name'].value_counts().index[1])
```

# Finding the total number of attacks according to their types

```python
attack_types = df1['Attack Type'].value_counts()
attack_types
```

# Finding the attack type with most attacks

```python
attack_types.plot(kind='bar',color='black',figsize = (17,10))
plt.ylabel('Number of Attacks')
plt.show()

print("Most attack type :",df1['Attack Type'].value_counts().idxmax())
```
